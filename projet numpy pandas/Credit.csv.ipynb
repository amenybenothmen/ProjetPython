##### b. Créer et supprimer  une colonne 

###### Création

```python
#telecharger de nouveau notre data frame 
credit = pd.read_csv("Credit.csv")
credit.head()
```

```python
# Créez une colonne duree à partir d'une colonne existante 
credit['duree']=credit['duration']+10
credit.head()
```

###### Suppression

Utiliser **drop**

```python
# Supprimer une colonne par la methode .drop
credit=credit.drop(columns='marital',axis=1)
credit.head()
```

```python
# Sélectionner les personnes qui ont des revenus > 10000
credit.loc[credit['balance']>10000, 'balance'].head()
```

```python
# Sélectionner les revenus des personnes qui ont des pret de logements 
credit.loc[credit['housing']=='yes', 'balance'].head()
```

```python
# calculer nombre de personnes approuvées
approuvee=credit.loc[credit['approval']=='yes', :]
dim= approuvee.shape
print(f"le nombre de personnes approuvees : {dim[0]}")
```

```python
# Sélectionner les personnes qui ont des revenus > 10000 et ont des logements 
credit.loc[(credit['balance']>10000) & (credit['housing']=='yes'), ['balance','housing']]
# calculer nombre de housing
housing=credit.loc[(credit['balance']>10000) & (credit['housing']=='yes'), ['balance','housing']]
dim= housing.shape
print(f"le nombre de housing et ont des revenus > 10000 : {dim[0]}")
```

```python
# Sélectionner les personnes qui ont des revenus > 10000 et ont des logements 
creditB = credit.loc[(credit['balance']>10000) & (credit['housing']=='yes'), :]
display(creditB)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Importing Seaborn for better aesthetics

credit = pd.read_csv('Credit.csv')

creditCourbe = credit.head(10)

duration = creditCourbe['duration']
balance = creditCourbe['balance']


# Create a scatter plot with a line connecting the points
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(duration, balance, marker='o', linestyle='-', color='b', label='Balance par rapport Duration')

# Add labels and title
plt.title('Balance par rapport Duration')
plt.xlabel('Duration')
plt.ylabel('Balance')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
